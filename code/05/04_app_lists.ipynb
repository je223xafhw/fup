````haskell
type ShowS = String -> String
````

````haskell
data Expr = Val Int
          | Expr :+: Expr
          | Expr :*: Expr

infixl 6 :+:
infixl 7 :*:
````

````haskell
instance Show Expr where
  showsPrec _ (Val n) = shows n
  --                       showParen Bool  ShowS -> ShowS
  --                       showParen Bool  ShowS -> ShowS
  --                                                                                     Int -> a -> ShowS
--                                                                                                    String -> String
  showsPrec p (e1 :+: e2) = showParen (p > 6) $ \s -> showsPrec 6 e1 $ showString " + " $ showsPrec 6 e2 s
  showsPrec p (e1 :*: e2) = showParen (p > 7) $ \s -> showsPrec 7 e1 $ showString " * " $ showsPrec 7 e2 s
  -- showsPrec p (e1 :+: e2) = showParen (p > 6) $ showsPrec 6 e1 . showString " + " . showsPrec 6 e2
  -- showsPrec p (e1 :*: e2) = showParen (p > 7) $ showsPrec 7 e1 . showString " * " . showsPrec 7 e2 
````

## ShowParen
<div style="width: 50%">
Die Funktion showParen nimmt zwei Argumente. Das erste Argument ist ein Boolescher Wert, der angibt, ob Klammern hinzugefügt werden sollen (True) oder nicht (False). Das zweite Argument ist eine ShowS-Funktion, die den inneren String repräsentiert, um den Klammern hinzugefügt werden sollen.
</div>

## `infixl 6 :+:`
<div style="width: 50%">
Im konkreten Fall bedeutet infixl 6 :+:, dass der :+:-Operator eine Fixität von 6 hat und linksassoziativ ist. Das heißt, wenn mehrere :+:-Operatoren in einem Ausdruck auftreten, wird der linkeste zuerst ausgewertet. Ein ähnliches Konzept gilt für den :*:-Operator, der eine Fixität von 7 und ebenfalls linksassoziativ ist.
Das ist also einfach nur 
</div>

```haskell 
let expr = Val 2 :+: Val 3 :*: Val 4
