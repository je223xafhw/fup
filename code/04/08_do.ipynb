```haskell
getNonEmptyLine :: IO String
getNonEmptyLine = do
  input  <- getLine
  result <- if null input 
              then do putStrLn "Please enter a non-empty string."
                      getNonEmptyLine
              else return input
  return result
```

```haskell
getNonEmptyLine :: IO String
getNonEmptyLine = getLine >>= 
    \input -> if null input then putStrLn "Please enter a non-empty string." >>= const getNonEmptyLine else return input
```
