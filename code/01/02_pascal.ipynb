Pascalsches Dreieck

````haskell
fac :: Int -> Int
fac 0 = 1
fac n = n * fac (n-1)
````

## Aufgabe 2.1
Binominalkoeffizient
$$
\binom nk = \frac{n!}{k! \cdot (n-k)!}
$$


````haskell
binom :: Int -> Int -> Int
binom n 0 = 1
binom n k  
    | n>=k = fac n `div` (fac k * fac(n-k))
    |otherwise = 0
````

## Aufgabe 2.2
Pascalsches Dreieck
- Bedingungen:
$
\begin{align}
x\;0&=1 \notag \\
x\;x&=1 \notag \\
x\;1&=x \notag \\
x\;x-1&=x \notag \\
\end{align}
$
- Algorithmus:
```
pascal(r p) = pascal((r-1) (p-1)) + pascal((r-1) p)
(6 3) = 
                  (5 2)             +        (5 3)
                    |                          |
            (4 1)   +       (4 2)        (4 2) + (4 3)
              |               |            |       |
        (3 0) + (3 1)   (3 1) + (3 2)     ...     ...
          |       |       |       |
          1       3       3      ...
```

````haskell
pascal :: Int -> Int -> Int
pascal r 0 = 1
pascal r 1 = r
pascal r p 
    | r == p = 1
    | p == (r-1) = r
    | r >= p = pascal (r-1) (p-1) + pascal (r-1) p
    | otherwise = 0
````

````haskell
print(pascal 6 4)
````
